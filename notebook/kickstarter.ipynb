```python
dfs = {
    2013: ttrpg_2013,
    2014: ttrpg_2014,
    2015: ttrpg_2015,
    2016: ttrpg_2016,
    2017: ttrpg_2017,
    2018: ttrpg_2018,
    2019: ttrpg_2019,
    2020: ttrpg_2020,
    2021: ttrpg_2021,
    2022: ttrpg_2022,
    2023: ttrpg_2023,
    2024: ttrpg_2024,
    2025: ttrpg_2025,}

ttrpg_all = pd.concat(
    [df.assign(year=year) for year, df in dfs.items()],
    ignore_index=True)
```

```python
ttrpg_all.info()
```

```python
cols = ["Description", "description", "Blurb"]

ttrpg_all[cols] = ttrpg_all[cols].replace(r"^\s*$", pd.NA, regex=True)

ttrpg_all["combined_description"] = (
    ttrpg_all[cols]
    .bfill(axis=1)
    .iloc[:, 0])
```

```python
ttrpg_all = ttrpg_all.drop(columns=["Description", "description", "Blurb"])
```

```python
ttrpg_all.info()
```

```python
ttrpg_all = ttrpg_all.loc[:, ~ttrpg_all.columns.str.startswith("Unnamed")]
```

```python
cols_to_drop = [
    "RPGGeek List Page",
    "Linked RPGs",
    "LevelUSD.1",
    "OSR Original",
    "OSR Scene",
    "Image Link",
    "Geeklist Code",
    'TrueLevelUSD ("Not Funded" = 0)',
    "FundedDate",
    "Funded Value",
    "OSR-Core",
    "OSR-Adjacent",
    "OSR Core",
    "OSR Adjacent",
    "RPGGeek List Item",
    "Funding Goal",
]

ttrpg_all = (
    ttrpg_all
    .loc[:, ~ttrpg_all.columns.str.startswith("Unnamed")]
    .drop(columns=cols_to_drop, errors="ignore")
)
```
