```python
dfs = {
    2013: ttrpg_2013,
    2014: ttrpg_2014,
    2015: ttrpg_2015,
    2016: ttrpg_2016,
    2017: ttrpg_2017,
    2018: ttrpg_2018,
    2019: ttrpg_2019,
    2020: ttrpg_2020,
    2021: ttrpg_2021,
    2022: ttrpg_2022,
    2023: ttrpg_2023,
    2024: ttrpg_2024,
    2025: ttrpg_2025,}

ttrpg_all = pd.concat(
    [df.assign(year=year) for year, df in dfs.items()],
    ignore_index=True)
```

```python
ttrpg_all.info()
```

```python
cols = ["Description", "description", "Blurb"]

ttrpg_all[cols] = ttrpg_all[cols].replace(r"^\s*$", pd.NA, regex=True)

ttrpg_all["combined_description"] = (
    ttrpg_all[cols]
    .bfill(axis=1)
    .iloc[:, 0])
```

```python
ttrpg_all = ttrpg_all.drop(columns=["Description", "description", "Blurb"])
```

```python
ttrpg_all.info()
```

```python
ttrpg_all = ttrpg_all.loc[:, ~ttrpg_all.columns.str.startswith("Unnamed")]
```

```python
cols_to_drop = [
    "RPGGeek List Page",
    "Linked RPGs",
    "LevelUSD.1",
    "OSR Original",
    "OSR Scene",
    "Image Link",
    "Geeklist Code",
    'TrueLevelUSD ("Not Funded" = 0)',
    "FundedDate",
    "Funded Value",
    "OSR-Core",
    "OSR-Adjacent",
    "OSR Core",
    "OSR Adjacent",
    "RPGGeek List Item",
    "Funding Goal",
]

ttrpg_all = (
    ttrpg_all
    .loc[:, ~ttrpg_all.columns.str.startswith("Unnamed")]
    .drop(columns=cols_to_drop, errors="ignore")
)
```

```python
ttrpg_all.info()
```

```python
ttrpg_all.head(10)
```

```python
ttrpg_all = ttrpg_all.drop(ttrpg_all.columns[[0, 1, 2, 3, 24]], axis=1)
```

```python
date_cols = ['funded_date', 'Ended?', 'Last Updated', 'Start Date']

for col in date_cols:
    ttrpg_all[col] = pd.to_datetime(ttrpg_all[col], errors='coerce')
```

```python
num_cols = ['Funding Level', 'LevelUSD', 'funded_goal']

for col in num_cols:
    ttrpg_all[col] = (
        ttrpg_all[col]
        .astype(str)
        .str.replace(r'[^\d.-]', '', regex=True)
        .replace('', pd.NA))
    ttrpg_all[col] = pd.to_numeric(ttrpg_all[col], errors='coerce')
```

```python
ttrpg_all['ExtractedFunding'] = (
    ttrpg_all['ExtractedFunding']
    .astype(str)
    .str.replace(r'[^\d.-]', '', regex=True)
    .astype(float))
```

```python
cols = ['Funded?', 'Cancelled?', 'Reached Goal?']

mapping = {
    'TRUE': 1.0, 'FALSE': 0.0,
    'YES': 1.0, 'NO': 0.0,
    'Y': 1.0, 'N': 0.0,
    True: 1.0, False: 0.0
}

for c in cols:
    ttrpg_all[c] = (
        ttrpg_all[c]
            .astype(str)
            .str.strip()
            .str.upper()
            .replace(mapping)
            .astype(float)
    )

ttrpg_all
```

1 = true, 0 = false

```python
success_col = 'Reached Goal?'

total_projects = ttrpg_all[success_col].notna().sum()
successful_projects = (ttrpg_all[success_col] == 1.0).sum()

success_rate = successful_projects / total_projects * 100

total_projects, successful_projects, success_rate
```

```python
yearly_success = (
    ttrpg_all
        .dropna(subset=[success_col])
        .groupby('year')[success_col]
        .mean() * 100
)

yearly_success
```

```python
median_yearly_success = yearly_success.median()
median_yearly_success
```

```python
ttrpg_all['is_5e'] = ttrpg_all['Non-5E: For RPG'].isna()

funded_by_category = (
    ttrpg_all
        .groupby('is_5e')['LevelUSD']
        .sum())

funded_by_category.index = funded_by_category.index.map({True: '5E', False: 'Non-5E'})
print(funded_by_category)
```

```python
currency_counts = ttrpg_all['Currency'].value_counts()

total_projects = len(ttrpg_all)

usd_pct = (currency_counts.get('USD', 0) / total_projects) * 100
other_pct = 100 - usd_pct

print(f"USD projects: {usd_pct:.2f}%")
print(f"Other currencies: {other_pct:.2f}%")
```

```python
funded_by_year_currency = ttrpg_all.groupby(['year', 'Currency'])['LevelUSD'].sum()
print(funded_by_year_currency)
```

```python
ttrpg_all['Type'] = ttrpg_all['Type'].replace(
    'https://rpggeek.com/rpg/1297/old-school-reference-and-index-compilation-osric', 
    pd.NA)
```

```python
ttrpg_all['type_filled'] = ttrpg_all['Type'].fillna('Unknown')

funding_by_year_type = ttrpg_all.groupby(['year', 'type_filled'])['LevelUSD'].sum()

funding_by_year_type_pivot = funding_by_year_type.unstack(fill_value=0)

funding_by_year_type_pivot = funding_by_year_type_pivot.applymap(lambda x: f"${x:,.0f}")

print(funding_by_year_type_pivot)
```

```python
ttrpg_all['type_filled'] = ttrpg_all['Type'].fillna('Unknown')

plt.figure(figsize=(12,6))
sns.barplot(
    data=ttrpg_all,
    x='year',
    y='LevelUSD',
    hue='type_filled',
    estimator=sum,   
    ci=None)

plt.title('TTRPG Funding per Year by Type')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Year')

formatter = FuncFormatter(lambda x, _: f"${x:,.0f}")
plt.gca().yaxis.set_major_formatter(formatter)

plt.legend(title='Type')
plt.xticks(rotation=45)
plt.show();
```

```python
labels = ['USD', 'Other currencies']
sizes = [usd_pct, other_pct]
colors = ['#1f77b4', '#ff7f0e'] 
explode = (0.1, 0)  

plt.figure(figsize=(6,6))
plt.pie(
    sizes, 
    labels=labels, 
    autopct='%1.1f%%',  
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True)
plt.title('Percentage of Projects in USD vs Other Currencies')
plt.show()
```

```python
yearly_success_df = yearly_success.reset_index()
yearly_success_df.columns = ['year', 'success_pct']

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_success_df, x='year', y='success_pct', marker='o')
plt.title('Yearly TTRPG Success Rate')
plt.ylabel('Success Rate (%)')
plt.xlabel('Year')
plt.ylim(0, 100)  
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
funding_long = funded_by_year_currency_pivot.reset_index().melt(
    id_vars='year',
    var_name='Currency',
    value_name='LevelUSD')

funding_long_non_usd = funding_long[funding_long['Currency'] != 'USD']

plt.figure(figsize=(12,6))
sns.barplot(
    data=funding_long_non_usd,
    x='year',
    y='LevelUSD',
    hue='Currency',
    estimator=sum,
    ci=None)

plt.title('TTRPG Funding per Year by Non-USD Currencies')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Year')

formatter = FuncFormatter(lambda x, _: f"${x:,.0f}")
plt.gca().yaxis.set_major_formatter(formatter)

plt.xticks(rotation=45)
plt.legend(title='Currency')
plt.show()
```

```python
funding_long = funded_by_year_currency_pivot.reset_index().melt(
    id_vars='year',
    var_name='Currency',
    value_name='LevelUSD')

funding_long = funding_long[funding_long['LevelUSD'] > 0]

plt.figure(figsize=(12,6))
sns.lineplot(
    data=funding_long,
    x='year',
    y='LevelUSD',
    hue='Currency',
    marker='o')

plt.title('TTRPG Funding per Year by Currency')
plt.ylabel('Total Funding (USD)')
plt.xlabel('Year')

formatter = FuncFormatter(lambda x, _: f"${x:,.0f}")
plt.gca().yaxis.set_major_formatter(formatter)

plt.xticks(rotation=45)
plt.legend(title='Currency', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
