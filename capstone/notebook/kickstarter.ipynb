```python
dfs = {
    2013: ttrpg_2013,
    2014: ttrpg_2014,
    2015: ttrpg_2015,
    2016: ttrpg_2016,
    2017: ttrpg_2017,
    2018: ttrpg_2018,
    2019: ttrpg_2019,
    2020: ttrpg_2020,
    2021: ttrpg_2021,
    2022: ttrpg_2022,
    2023: ttrpg_2023,
    2024: ttrpg_2024,
    2025: ttrpg_2025,}

ttrpg_all = pd.concat(
    [df.assign(year=year) for year, df in dfs.items()],
    ignore_index=True)
```

```python
cols = ["Description", "description", "Blurb"]

ttrpg_all[cols] = ttrpg_all[cols].replace(r"^\s*$", pd.NA, regex=True)

ttrpg_all["combined_description"] = (
    ttrpg_all[cols]
    .bfill(axis=1)
    .iloc[:, 0])
```

```python
ttrpg_all = ttrpg_all.drop(columns=["Description", "description", "Blurb"])
```

```python
cols_to_drop = [
    "RPGGeek List Page",
    "Linked RPGs",
    "LevelUSD.1",
    "OSR Original",
    "OSR Scene",
    "Image Link",
    "Geeklist Code",
    'TrueLevelUSD ("Not Funded" = 0)',
    "FundedDate",
    "Funded Value",
    "OSR-Core",
    "OSR-Adjacent",
    "OSR Core",
    "OSR Adjacent",
    "RPGGeek List Item",
    "Funding Goal",
]

ttrpg_all = (
    ttrpg_all
    .loc[:, ~ttrpg_all.columns.str.startswith("Unnamed")]
    .drop(columns=cols_to_drop, errors="ignore")
)
```

```python
num_cols = ['Funding Level', 'LevelUSD', 'funded_goal']

for col in num_cols:
    ttrpg_all[col] = (
        ttrpg_all[col]
        .astype(str)
        .str.replace(r'[^\d.-]', '', regex=True)
        .replace('', pd.NA))
    ttrpg_all[col] = pd.to_numeric(ttrpg_all[col], errors='coerce')
```

```python
cols = ['Funded?', 'Cancelled?', 'Reached Goal?']

mapping = {
    'TRUE': 1.0, 'FALSE': 0.0,
    'YES': 1.0, 'NO': 0.0,
    'Y': 1.0, 'N': 0.0,
    True: 1.0, False: 0.0
}

for c in cols:
    ttrpg_all[c] = (
        ttrpg_all[c]
            .astype(str)
            .str.strip()
            .str.upper()
            .replace(mapping)
            .astype(float)
    )

ttrpg_all
```

```python
success_col = 'Reached Goal?'

total_projects = ttrpg_all[success_col].notna().sum()
successful_projects = (ttrpg_all[success_col] == 1.0).sum()

success_rate = successful_projects / total_projects * 100

total_projects, successful_projects, success_rate
```

```python
currency_cols = ['Funding Level', 'Funding Level.1', 'FundingLevelUSD']

def to_numeric_currency(x):
    if pd.isnull(x):
        return float('nan')
   
    cleaned = ''.join(c for c in str(x) if c.isdigit() or c in ['.', '-'])
    try:
        return float(cleaned)
    except:
        return float('nan')


for col in currency_cols:
    gamefound[col] = gamefound[col].apply(to_numeric_currency)
```
